```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取数据
df = pd.read_csv(r'D:\EDA - iPhone Resale Pricing & Market Signals\ecommerce_iphone_resale_market_intelligence_usa_2026.csv')

# 2. 计算每个型号的平均价格
price_summary = df.groupby('model_family')['price'].mean().sort_values()

# 3. 定义苹果风格颜色
BG_DARK = '#0D1117'
BG_CARD = '#161B22'
TEXT_PRIMARY = '#E6EDF3'
TEXT_SECONDARY = '#8B949E'
GRID_COLOR = '#21262D'
MY_COLOR = "#f0edee"
APPLE_BLUE = '#007AFF'

fig,ax = plt.subplots(figsize=(10, 6))
#fig画一个图，ax选定一个区域，figsize说明图的宽是10，长度是6

fig.patch.set_facecolor(BG_DARK)
#fig.patch是整张图的最外层背景，设置为苹果风格的深色背景，set_facecolor设置背景色
ax.set_facecolor(BG_CARD)
#ax是中间画图的区域的背景，设置为苹果风格的卡片背景，set_facecolor设置背景色，中间比外面稍亮一点，看起来更高级
bars = ax.barh(price_summary.index, price_summary.values, color=APPLE_BLUE)
#ax.barh是画水平方向的柱状图，第一个参数是纵轴上的内容，也就是苹果手机的型号，第二个参数是横轴上的内容，也就是平均价格，设置颜色为苹果风格的蓝色
ax.set_title("Average Price by Model Family", color=TEXT_PRIMARY,fontsize=14,loc='center')
#设置图标题，颜色是标题颜色，大小14，靠中
ax.set_xlabel("Average Price ($)", color=MY_COLOR,fontsize=12)
#设置x轴标签，颜色是标签颜色，大小12
ax.set_ylabel("Model Family", color=MY_COLOR,fontsize=12)
#设置y轴标签，颜色是标签颜色，大小12
ax.tick_params(colors=MY_COLOR)
#设置x轴和y轴的刻度颜色,刻度颜色就是x轴上的数字，和y轴上的型号名
ax.spines['top'].set_visible(False)
#设置顶部轴线不可见
ax.spines['right'].set_visible(False)
#设置右侧轴线不可见
ax.spines["left"].set_color(APPLE_BLUE)
#设置左侧轴线颜色为网格颜色
ax.spines["bottom"].set_color(APPLE_BLUE)
#设置底部轴线颜色为网格颜色
ax.grid(axis='x', color=GRID_COLOR, linewidth=0.5,linestyle='--', alpha=0.3)
#设置x轴网格线为虚线，透明度为0.3
ax.grid(axis='y', visible=False)
#设置y轴方向网格线不可见

for bar in bars:
    width = bar.get_width()
    ax.text(
        width + 10,                  # 文字放在柱子右边一点，bar.get_width()是柱子从左往右地高度，height是从上往下，width是从左往右
        bar.get_y() + bar.get_height() / 2,   # 文字放在柱子中间高度，bar.get_y()是柱子底部地位置，bar.get_height()是柱子地垂直（从下往上）
        f'${width:.0f}',             # 显示成美元数字，文字内容
        va='center',                 # 垂直居中
        ha='left',                   # 文字从左开始写
        color=TEXT_PRIMARY,
        fontsize=10
    )

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches

df = pd.read_csv('ecommerce_iphone_resale_market_intelligence_usa_2026.csv')

model_counts = df['model_family'].value_counts().sort_values()
is_pro = model_counts.index.str.contains('Pro')
is_pro
model_counts
```

```python

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('#F7F9FC')
ax.set_facecolor('#F7F9FC')

y = range(len(model_counts))  #生成一串y的位置（纵坐标），从下往上画

for i, count in enumerate(model_counts.values):
    color = '#0B1F3A' if is_pro[i] else '#4E79A7'  #线的颜色
    ax.hlines(y[i], xmin = 0, xmax = count, color=color, linewidth=0.8) #线的位置，线的左端点，线的右端点

for i, count in enumerate(model_counts.values):
    color = '#0B1F3A' if is_pro[i] else '#4E79A7'
    ax.scatter(count, y[i], color=color, s=100, alpha=0.8) #线的横坐标，纵坐标

for i,count in enumerate(model_counts.values):
    ax.text(count+8, y[i]-0.35, str(count), fontsize=12, color='black') #线的横坐标，纵坐标，文本内容

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

outer_border = patches.Rectangle(
    (0.01, 0.01), 0.98, 0.98, 
    transform = fig.transFigure,
    fill=False, 
    edgecolor='black', 
    linewidth=1)
fig.add_artist(outer_border)

title_box = patches.Rectangle(
    (0.01, 0.94), 0.98, 0.05, 
    transform = fig.transFigure,
    fill=True, 
    edgecolor='black', 
    linewidth=1)
fig.add_artist(title_box)

fig.text(
    0.5, 0.963,
    'iPhone Model Count',
    fontsize=14,
    color='white',
    ha='center',
    va='center',
    fontweight='bold'
)

ax.set_yticks(list(y))
ax.set_yticklabels(model_counts.index)

ax.set_xlabel('Count')

pro_patch = patches.Patch(
    color='#0B1F3A',
    label='Pro/Pro Max'
)

standard_patch = patches.Patch(
    color='#4E79A7',
    label='Standard/Plus/Mini'
)

legend = ax.legend(
    handles=(pro_patch, standard_patch),
    loc='lower right',
    fontsize=12,
    frameon=True,
    facecolor='white',
    edgecolor='black',
)

fig.subplots_adjust(left=0.16,right=0.95,bottom=0.15)


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches

df = pd.read_csv('ecommerce_iphone_resale_market_intelligence_usa_2026.csv')

gen_pro = df.groupby(['generation_number', 'is_pro']).size().unstack(fill_value=0)
F_pro = (gen_pro[False]/(gen_pro[True]+gen_pro[False])).round(2)
F_standard = (gen_pro[True]/(gen_pro[True]+gen_pro[False])).round(2)

fig,axes = plt.subplots(1,2,figsize=(12,5))
fig.patch.set_facecolor('#F7F7F5')
ax1= axes[0]
ax2= axes[1]

# ---------------- 左图：数量堆叠图 ----------------

standard = gen_pro[False]
pro = gen_pro[True]
x = gen_pro.index.astype(int).astype(str)

ax1.set_facecolor('#F7F7F5')

bars1 = ax1.bar(x, standard, label='Standard', color='#4E79A7')
bars2 = ax1.bar(x, pro, bottom=standard, label='Pro', color='#0B1F3A')


ax1.set_xlabel('Generation Number', color='#4A4A4A')
ax1.set_ylabel('Sample Count', color='#4A4A4A')
ax1.tick_params(colors='#4A4A4A')

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_color('#B8B8B8')
ax1.spines['bottom'].set_color('#B8B8B8')


standard_patch = patches.Patch(color='#4E79A7', label='Standard')
pro_patch = patches.Patch(color='#0B1F3A', label='Pro')
legend1 = ax1.legend(handles=[standard_patch, pro_patch], loc='upper right', fontsize=8)

# 给柱子中间加数字
for i, bar in enumerate(bars1):
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height / 2,
        f'{int(height)}',
        ha='center',
        va='center',
        color='white',
        fontsize=8,
        fontweight='bold'
    )

for i, bar in enumerate(bars2):
    height = bar.get_height()
    bottom_value = standard.iloc[i]
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        bottom_value + height / 2,
        f'{int(height)}',
        ha='center',
        va='center',
        color='white',
        fontsize=8,
        fontweight='bold'
    )
    y=bar.get_y()


# ---------------- 右图：比例横向堆叠图 ----------------
ax2.set_facecolor('#F7F7F5')
bars3 = ax2.barh(x, F_pro, label='Pro', color='#4E79A7')
bars4 = ax2.barh(x, F_standard, left=F_pro, label='Standard', color='#0B1F3A')


ax2.set_xlabel('Ratio', color='#4A4A4A')
ax2.set_ylabel('Generation Number', color='#4A4A4A')
ax2.tick_params(colors='#4A4A4A')

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_color('#B8B8B8')
ax2.spines['bottom'].set_color('#B8B8B8')

ax2.legend(handles=[pro_patch, standard_patch], loc='upper right', fontsize=8)


# 给柱子中间加数字
for i, bar in enumerate(bars3):
    height = bar.get_height()
    ax2.text(
        bar.get_width()/2,
        bar.get_y() + height/2,
        f'{bar.get_width():.0%}',
        ha='center',
        va='center',
        color='white',
        fontsize=8,
        fontweight='bold'
    )


for i, bar in enumerate(bars4):
    height = bar.get_height()
    left_value = F_pro.iloc[i]
    ax2.text(
        bar.get_width() / 2 + left_value,
        bar.get_y() + height / 2,
        f'{bar.get_width():.0%}',
        ha='center',
        va='center',
        color='white',
        fontsize=8,
        fontweight='bold'
    )

outer_figure1 = patches.Rectangle(
    (0.06, 0.01), 0.43, 0.98,    #0.43是从0.06的基础上再往右0.43
    transform = fig.transFigure,
    fill=False, 
    edgecolor='black', 
    linewidth=1)
fig.add_artist(outer_figure1)

outer_figure2 = patches.Rectangle(
    (0.49, 0.01), 0.44, 0.98, 
    transform = fig.transFigure,
    fill=False, 
    edgecolor='black', 
    linewidth=1)
fig.add_artist(outer_figure2)


title_box1 = patches.Rectangle(
    (0.06, 0.94), 0.43, 0.05, 
    transform = fig.transFigure,
    fill=True, 
    edgecolor='black', 
    linewidth=1)
fig.add_artist(title_box1)

title_box2 = patches.Rectangle(
    (0.49, 0.94), 0.44, 0.05, 
    transform = fig.transFigure,
    fill=True, 
    edgecolor='black', 
    linewidth=1)
fig.add_artist(title_box2)

fig.text(
    0.27, 0.965,
    'Listing by generation',
    fontsize=12,
    color='white',
    ha='center',
    va='center',
    fontweight='bold'
)

fig.text(
    0.27, 0.92,
    'absolute count —— Standard vs Pro vier',
    fontsize=10,
    color='black',
    ha='center',
    va='center',    
)

fig.text(
    0.72, 0.965,
    'Pro vs Standard Mix',
    fontsize=12,
    color='white',
    ha='center',
    va='center',    
    fontweight='bold'
)
fig.text(
    0.72, 0.92,
    'Ratio of Pro/Standard by Generation Number',
    fontsize=10,
    color='black',
    ha='center',
    va='center',    

)

plt.show()
```

```python
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt

seller_stats = df.groupby('seller').agg(
    listings=('price', 'count'),
    median_price=('price', 'median'),
    total_value=('price', 'sum'),
    models=('model_family', 'nunique'),
    conditions=('condition', 'nunique')
).sort_values('listings', ascending=False)

fig, ax = plt.subplots(figsize=(6, 4))
fig.patch.set_facecolor('#F7F7F5')
ax.set_facecolor('#F7F7F5')

cum_listings = seller_stats['listings'].cumsum() / seller_stats['listings'].sum() * 100
cum_value = seller_stats['total_value'].cumsum() / seller_stats['total_value'].sum() * 100
x_pct = np.arange(1, 1 + len(cum_listings)) / len(cum_listings) * 100

ax.plot(x_pct, cum_listings.values, color="#0B1F3A", linewidth=2.5, label='Listing Volume')
ax.plot(x_pct, cum_value.values, color="#4E79A7", linewidth=2.5, linestyle='--', label='Total Value')
ax.plot([0, 100], [0, 100], color="#A8A8A8", linestyle=':', linewidth=1.2, label="Perfect equality")

ax.set_xlabel('Percentile', color='#4A4A4A')
ax.set_ylabel('Cumulative Share', color='#4A4A4A')
ax.tick_params(colors='#4A4A4A')

ax.axhline(80, color='#7A8694', linestyle='--', alpha=0.6)

ax.spines['left'].set_color('#B8B8B8')
ax.spines['bottom'].set_color('#B8B8B8')
ax.spines['top'].set_color('#B8B8B8')
ax.spines['right'].set_color('#B8B8B8')


outer_figure = patches.Rectangle(
    (0, 0), 1, 1.1,
    transform=fig.transFigure,
    fill=False,
    edgecolor='#7A8694',
    linewidth=1
)
fig.add_artist(outer_figure)

title_figure = patches.Rectangle(
    (0, 0.95), 1, 0.15,
    transform=fig.transFigure,
    facecolor='#0B1F3A',
    edgecolor='#0B1F3A',
    linewidth=1
)
fig.add_artist(title_figure)

fig.text(
    0.3, 1,
    'Seller Statistics',
    color="white",
    transform=fig.transFigure,
    fontsize=20,
    fontweight='bold'
)

legend = ax.legend(
    facecolor='white',
    edgecolor='#D9D9D9',
    fontsize=9,
    frameon=True
)

for text in legend.get_texts():
    text.set_color('#4A4A4A')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.stats import gaussian_kde

df = pd.read_csv('ecommerce_iphone_resale_market_intelligence_usa_2026.csv')
bins = np.linspace(0, 2000, 80)  # 80个bin，每个bin的间距相等，包括0和包括2000
fig,ax = plt.subplots(figsize=(12, 6))
hist = ax.hist(df["price"],bins=bins,color="steelblue",alpha=0.6)

price_clip = df["price"][df["price"]<=2000]
kde = gaussian_kde(price_clip,bw_method=0.08)
x_kde = np.linspace(0, 2000, 500)
y_kde = kde(x_kde)

y_kde = y_kde/y_kde.max()*hist[0].max()
ax.plot(x_kde,y_kde,color="blue",linestyle="--",linewidth=1,alpha=0.6,label="KDE")

mean_price = df["price"].mean()
median_price = df["price"].median()
ax.axvline(mean_price,color="red",linestyle="--",linewidth=1,label="Mean")
ax.axvline(median_price,color="green",linestyle="--",linewidth=1,label="Median")
ax.legend()
ax.set_xlabel("Price")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of iPhone Resale Prices")
ax.text(520, 100, f"Mean: {mean_price:.2f}", color="red",fontsize=7)
ax.text(330, 250, f"Median: {median_price:.2f}", color="green",fontsize=7)
price_outfigure = (df["price"]>2000).sum()
ax.text(1700, 150, f"Outfigure number: {price_outfigure:.0f}\n            ————>", color="orange",fontsize=9)
```
